```python
with open("11input.txt") as file:
    s = [list(l) for l in file.read().split("\n")]

def expansion_of_space_distances(expansion, space):
    expansion -= 1 
    index_large = 0
    galaxy = 1
    galaxy_dict = {}

    for index_real in range(len(space)):
        line = space[index_real]
        if "#" not in line:
            index_large += expansion
        else:
            for char in range(len(line)):
                if line[char] == "#":
                    space[index_real][char] = str(galaxy)
                    galaxy_dict[galaxy] = [index_large, None] 
                    galaxy += 1
        index_large += 1

    index_large = 0
    for index_real in range(len(space[0])):  
        column = [sublist[index_real] for sublist in space] 
        if all(element == "." for element in column):
            index_large += expansion
        else:
            for row in range(len(column)):
                if column[row].isdigit():
                    galaxy_dict[int(column[row])][1] = index_large
        index_large += 1

    def shortest_path(dict_of_galaxies, element1, element2):
        place1 = dict_of_galaxies[int(element1)]
        place2 = dict_of_galaxies[int(element2)]
        cols = abs(place1[1] - place2[1])
        lines = abs(place1[0] - place2[0])
        return cols + lines

    path = 0
    for i in range(1, galaxy):
        for j in range(i, galaxy):
            path += shortest_path(galaxy_dict, str(i), str(j))
            
    return path

print(expansion_of_space_distances(2, [row.copy() for row in s]))
print(expansion_of_space_distances(1000000, [row.copy() for row in s]))
```
