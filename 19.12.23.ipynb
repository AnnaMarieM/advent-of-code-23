```python
#with open("19.txt") as file:
with open("19input.txt") as file:
    workflow, parts = [l.split("\n") for l in file.read().split("\n\n")]

# Part 1
workflows = {l.split("{")[0]: l.split("{")[1][:-1] for l in workflow}
for key, w in workflows.items():
    w_dict = []
    for instruction in w.split(","):
        if "<" in instruction or ">" in instruction:
            w_dict.extend([instruction[0], (instruction[1:].split(":")[0], instruction[1:].split(":")[1])])
        else:
            w_dict.extend([instruction])
    workflows[key] = w_dict
    
def process(part, name_of_workflow):
    instructions = workflows[name_of_workflow]
    for i, name in enumerate(instructions[::2]):
        if name in part.keys():
            if eval(part[name] + instructions[2*i + 1][0]):
                if instructions[2*i + 1][1] == "R":
                    return
                elif instructions[2*i + 1][1] == "A":
                    return [int(p) for p in part.values()]
                else:
                    return process(part, instructions[2*i + 1][1])
    if instructions[-1] == "R":
        return
    elif instructions[-1] == "A":
        return [int(p) for p in part.values()]
    else:
        return process(part, instructions[-1])

parts = [{p.split("=")[0]: p.split("=")[1] for p in part} for part in [part.lstrip("{").rstrip("}").split(",") for part in parts]]

suma = 0
for part in parts:
    outcome = process(part, "in")
    if outcome != None:
        suma += sum(outcome)
        
print(suma) 

# Part 2
workflows = {l.split("{")[0]: l.split("{")[1][:-1] for l in workflow}

def accepted_intervals(workflow):
    if type(workflow) == str:
        workflow = workflows[workflow].split(",")
    instruction = workflow[0]
    if instruction == "R":
        return []
    if instruction == "A":
        return [((1, 4000), (1, 4000), (1, 4000), (1, 4000))]
    if ":" not in instruction:
        return accepted_intervals(instruction)
        
    xmas_condition = instruction.split(":")[0]
    if ">" in xmas_condition:
        xmas_condition_true = ranges(xmas_condition[0], True, int(xmas_condition[2:]), accepted_intervals([instruction.split(":")[1]]))
        xmas_condition_false = ranges(xmas_condition[0], False, int(xmas_condition[2:]) + 1, accepted_intervals(workflow[1:]))

    elif "<" in xmas_condition:
        xmas_condition_true = ranges(xmas_condition[0], False, int(xmas_condition[2:]), accepted_intervals([instruction.split(":")[1]]))
        xmas_condition_false = ranges(xmas_condition[0], True, int(xmas_condition[2:]) - 1, accepted_intervals(workflow[1:]))

    return xmas_condition_true + xmas_condition_false

def ranges(xmas, greater, value, ranges):
    ranges2 = []
    for interval in ranges:
        interval = list(interval)
        lower_bound, upper_bound = interval['xmas'.index(xmas)]
        if greater:
            lower_bound = max(lower_bound, value + 1)
        else:
            upper_bound = min(upper_bound, value - 1)
        if lower_bound <= upper_bound:
            interval['xmas'.index(xmas)] = (lower_bound, upper_bound)
            ranges2.append(tuple(interval))
    return ranges2

suma2 = 0
for xmas_intervals in accepted_intervals('in'):
    combinations = 1
    for lower_bound, upper_bound in xmas_intervals:
        combinations *= upper_bound + 1 - lower_bound
    suma2 += combinations
    
print(suma2)
```
